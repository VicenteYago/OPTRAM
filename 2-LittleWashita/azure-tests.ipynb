## 1 - Connect to workspace

https://docs.microsoft.com/en-us/azure/machine-learning/how-to-manage-workspace?tabs=python#connect-to-a-workspace

```python
from azureml.core.authentication import InteractiveLoginAuthentication
from azureml.core import Workspace
from azureml.core import Dataset
from azureml.data.datapath import DataPath



interactive_auth = InteractiveLoginAuthentication(tenant_id="a2b6ab86-86a1-4e35-a3ad-09be0ec3fe3d")
ws = Workspace.from_config(auth=interactive_auth)
print(ws)
```

## 2- Create Azure Files dataset

https://docs.microsoft.com/en-us/azure/machine-learning/how-to-datastore?tabs=cli-identity-based-access%2Ccli-adls-identity-based-access%2Csdk-azfiles-accountkey%2Ccli-adlsgen1-identity-based-access#create-an-azure-files-datastore


```python
datastore = ws.get_default_datastore()
# Uploading local directory and creating a FileDataset
dataset = Dataset.File.upload_directory(
    src_dir="./sensor_data/", target=DataPath(datastore, "/sensor_data_washita/"), show_progress=True
)
```

```python

# You can register the datastore created by uploading the directory
dataset.register(
    workspace=ws,
    name="ds_from_directory",
    description="This dataset was creating by uploading a directory",
)
```
